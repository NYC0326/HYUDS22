```python
# This is a guess the number game.
import random

print('Hello! What is your name?')
myName = input()

number = random.randint(1,20)
print('Well, ' + myName + ', I am thinking of a number between 1 and 20.')

guess = 0
while guess != number:
    print('Take a guess.')
    guess = input()
    guess = int(guess)
    
    if guess < number:
        print('Your guess is too low.')
    if guess > number:
        print('Your guess is too high.')
    if guess == number:
        print('Good job, ' + myName + '! You guessed my number!')
```

```python
# This is a guess the number game.
import random

print('Hello! What is your name?')
myName = input()

number = random.randint(1,20)
print('Well, ' + myName + ', I am thinking of a number between 1 and 20.')

guess = 0
guessesTaken = 0
while guess != number:
    guessesTaken = guessesTaken + 1
    print('Take a guess.')
    guess = input()
    guess = int(guess)
    
    if guess < number:
        print('Your guess is too low.')
    if guess > number:
        print('Your guess is too high.')
    if guess == number:
        guessesTaken = str(guessesTaken)
        print('Good job, ' + myName + '! You guessed my number!' + guessesTaken + 'guesses!')        
```

### Formatting

```python
print('integer: %d, string: %s, float: %f' %(100, 'hi', 3.14))
```

```python
print('My name is %s' % 'Hong')
print('My name is (%7s)' % 'Hong')
print('My name is (%-7s)' % 'Hong')
print('My name is (%7s)' % 'Hong Gildong')
```

```python
print('code: %d' % 777)
print('code: %4d' % 777)
print('code: %-4d' % 777)
print('code: %04d' % 777)
print('code: %02d' % 777)
```

```python
print('pi: %f' % 3.141592)
print('pi: %.2f' % 3.141592)
print('pi: %.4f' % 3.141592)
print('pi: %10.2f' % 3.141592)
print('pi: %010.2f' % 3.141592)
```

```python
print('integer: {}, string: {}, float: {}'.format(100, 'hi', 3.14))
```

```python
print('pi: {}'.format(3.141592))
print('pi: {:.2f}'.format(3.141592))
print('pi: {:.4f}'.format(3.141592))
print('max:{a:.2f}, min:{b:.2f}'.format(a=2.222, b=1.1111))
print('max:{b:.2f}, min:{a:.2f}'.format(a=0.222, b=1.1111))
```

```python
radius = 7
pi = 3.141592
print(f'area = {pi*radius*radius}')

area = pi*radius*radius

print(f'area = {area:.2f}')
print(f'area = {area:.4f}')
```
